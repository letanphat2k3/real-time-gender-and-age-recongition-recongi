```python
BATCH_SIZE = 64
EPOCHS = 80
TARGET_SIZE = (128,128)

MIN_DEGREE = 15
MAX_DEGREE = 45
```

```python
def process_image(image_path, degrees, filename):
    image = Image.open(image_path)
    # image = image.rotate(degrees, expand = True, resample = Image.NEAREST)
    image = image.resize(TARGET_SIZE)
    image = image.transpose(Image.FLIP_LEFT_RIGHT)
    image.save(filename)

save_rotated_folder = r"D:\ML\mainAgeDec\UTKFace\vertical"

image_folder = r"D:\ML\mainAgeDec\UTKFace\origin_images"
images_files = os.path.join(image_folder, '*.jpg')

for image_path in tqdm(glob.glob(images_files), unit = "images"):
    # sign = random.choice([-1, 1])
    # degree = sign * random.randint(MIN_DEGREE, MAX_DEGREE)

    filename = image_path.split('\\')[-1]
    filename = filename.split(".jpg")[0]
    process_image(image_path, 0, save_rotated_folder + f"\{filename}_vertical.jpg")
```

```python
image_folder = r"D:\ML\mainAgeDec\UTKFace\data"
images_files = os.path.join(image_folder, '*.jpg')

for image_path in tqdm(glob.glob(images_files), unit = "images"):
    file_name_parts = os.path.splitext(image_path)
    # Check if the filename contains "rotated"
    if "ROTATED_" in file_name_parts[0]:
        # Rename the file by moving "rotated" to the end of the filename
        new_file_name = file_name_parts[0].replace("ROTATED_", "") + "_rotated" + file_name_parts[1]
        os.rename(image_path, new_file_name)
        print("File renamed successfully.")
    else:
        print("No 'rotated' part found in the filename.")
```
